fonksiyonun kaç tane parametresi varsa daima o kadar input vermemiz gerekir aksi taktirde programımız çalışmaz bizlere hata verir.

```python
def say_hello(name,surname):
    print(f'{name} {surname} welcome to new world!')
say_hello('John','Walter')
```

```python
say_hello('Boran')
```

burada gördüğümüz gibi 1 tane parametreyi eksik verdik ve hemen yanıt olarak hata kısmında bunu bizlere bildiriyor.

şimdi bir diğer önemli ayrıntı fonksiyonlar konusunda return ile programın outputunu çağırıldığı yere döndürmemizi sağlar.

```python
print(say_hello('Boran','Dabak'))
```

burada gördüğümüz üzere sonucu yazdırdı fakat ardından None değerini döndürdü bunun olmaması için return etmemiz gerekir.

```python
def absolute_value(num):
    if num >= 0:
        return num
    else:
        return -num


print(absolute_value(2))

print(absolute_value(-4))
```

fonksiyonların genel olarak çalışma şekli yukarıda gösterildiği gibidir.


şimdi burada diğer dikkat etmemiz gereken konu şu global value ve local value.

Global value: mesela 3.14 pi sayısını tanımlar dünyanın her yerinde böyledir. işte bu da programa tanımladığımız değişkenin tüm program boyunca sabit bir değerinin olması anlamına gelir.

Local value: mesela her ülkenin kendilerine göre kuralları olması local diyebiliiriz çünkü o ülkenin şartlarına göre değişkenlik gösterir. Local value tam olarak böyledir aslında bir değişken tanımlarız ve sonrasında programın gidiş yönüne göre değeri değişebilir

```python
def my_func():
    x = 10
    print("Value inside function:",x)

x = 20
my_func()
print("Value outside function:",x)
```

```python
def greet(name, msg="Good morning!"):

    print("Hello", name + ', ' + msg)


greet("Kate")
greet("Bruce", "How do you do?")
```

bazen bir fonksiyon oluşturduğumuz zaman kaç tane parametre alabileceğini bilemeyebiliriz 100 tane olabilir 1 tane olabilir böyle durumlarda yapmamız gereken bir tane parametre yazıp başına * koymak

```python
def greet(*names):

    for name in names:
        print("Hello", name)


greet("Monica", "Luke", "Steve", "John")
```
